```python
#Linear Algebra for picture SVD(Single Value Decomposition)
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
#import picture
pic = Image.open('Einstein_tongue.jpg')

#Let's have a look
plt.imshow(pic)
plt.show()

#we need to convert it to Floating-Point precision
pic = np.array(pic)
plt.imshow(pic,cmap='gray')
plt.show()
```

```python
#SVD(Single Value Decomposition)

#do the SVD, Note: You'll understand this decomposition by the
#end of course! Don't worry if it seems mysterious now!

U,S,V = np.linalg.svd(pic)

#plot the spectrum
plt.plot(S,'s-')
plt.xlim([0,100])
plt.xlabel('Component number')
plt.ylabel('singular value (\sigma)')
plt.show()
```

```python
#reconstruct the image based on some components
#list the components you want to use for the reconstruction
comps1 = np.arange(0,5)
comps2 = np.arange(20,151)
comps3 = np.arange(0,151)
comps4 = np.arange(0,20)

#reconstruct the low-rank version of the picture
reconpic1 = U[:,comps1]@np.diag(S[comps1])@V[comps1,:]
reconpic2 = U[:,comps2]@np.diag(S[comps2])@V[comps2,:]
reconpic3 = U[:,comps3]@np.diag(S[comps3])@V[comps3,:]
reconpic4 = U[:,comps4]@np.diag(S[comps4])@V[comps4,:]

#show the original and reconstructed pictures for comparison
#1)show the original picture
plt.figure(figsize=(20,30))
plt.subplot(1,6,1)
plt.imshow(pic,cmap='gray')
plt.title('original')
plt.axis('off')

#2)show reconstructed picture with [0:5] components
plt.subplot(1,6,2)
plt.imshow(reconpic1,cmap='gray')
plt.title('components %s-%s'%(comps1[0],comps1[-1]))
plt.axis('off')
plt.show()

#3)show reconstructed picture with [20:151] components
plt.subplot(1,6,3)
plt.imshow(reconpic2,cmap='gray')
plt.title('components %s-%s'%(comps2[0],comps2[-1]))
plt.axis('off')
plt.show()

#4)show reconstructed picture with [0:151] components
plt.subplot(1,6,4)
plt.imshow(reconpic3,cmap='gray')
plt.title('components %s-%s'%(comps3[0],comps3[-1]))
plt.axis('off')
plt.show()

#5)show reconstructed picture with [0:20] components
plt.subplot(1,6,5)
plt.imshow(reconpic4,cmap='gray')
plt.title('components %s-%s'%(comps4[0],comps4[-1]))
plt.axis('off')
plt.show()

#Aren't you SUPER-curios to know what all of this means and 
#why it works!!??! You're going to learn all about this in the course
```

## VIDEO: vector vector multiplication: the dot product

```python
## many ways to compute the dot product

v1 = np.array([1,2,3,4,5,6])
v2 = np.array([0,-4,-3,6,5,8])

#method1
dp1 = sum(np.multiply(v1,v2))

#method2
dp2 = np.dot(v1,v2)

#method3
dp3 = np.matmul(v1,v2)

#method4
dp4 = 0 #intialize

#loop over elements
for i in range(0,len(v1)):
    #multiply corresponding elements and sum
    dp4 = dp4 + v1[i]*v2[i]

print(dp1,dp2,dp3,dp4)
```

## VIDEO: The dot product from a geometric prespective

```python
#two vectors
v1 = np.array([2,4,-3])
v2 = np.array([0,-3,-3])

#compute the angle (radians) between two vectors
ang = np.arccos(np.dot(v1,v2)/(np.linalg.norm(v1)*np.linalg.norm(v2)))

#draw them
fig = plt.figure()
ax = fig.add_subplot(111,projection='3d')
ax.plot([0,v1[0]],[0,v1[1]],[0,v1[2]],'b')
ax.plot([0,v2[0]],[0,v2[1]],[0,v2[2]],'r')

plt.axis((-6,6,-6,6,-6,6))
plt.title('Angle between vectors: %s rad.'%ang)
plt.show()
```

```python
##equivalence of algebric and geometric dot product formulas

#two vectors
v1 = np.array([2,4,-3])
v2 = np.array([0,-3,-3])

#algebric
dp_a = np.dot(v1,v2)

#geometric
dp_g = np.linalg.norm(v1)*np.linalg.norm(v2)*np.cos(ang)

#print dot product to command
print(dp_a)
print(dp_g)
```

```python
## part1: implement MAI(Matrix Assymetry Index)
def MAI(A):
    Aanti = (A-A.T)/2
    mai = np.linalg.norm(Aanti)/np.linalg.norm(A)
    return mai

## part2: compute MAI for symmetric, skew-symmetric and random matrix
A = np.random.rand(5,5)
A = (A+A.T)/2

##part3: formula for mixing skew/symmetric matrices
p = 0
A = np.random.randn(5,5)
B = (1-p)*(A+A.T)/2 + p*(A-A.T)/2

##part4: test on random matrices
ps = np.linspace(0,1,100)
mai = np.zeros(len(ps))
for i in range(len(ps)):
    #create a matrix
    p = ps[i]
    A = np.random.randn(5,5)
    B = (1-p)*((A+A.T)/2) + p*((A-A.T)/2)
    #compute and store the MAI
    mai[i] = MAI(B)

plt.plot(ps,mai,'s-')
plt.show()
```

## Code challenge: implement the MCA algorithm

```python
#MCA algorithm in code!!
#then compute using inv() and compare to MCA
m = 4
A = np.random.randn(m,m)

##minors matrix
minors = np.zeros((m,m))
H = np.zeros((m,m))
for i in range(m):
    for j in range(m):
        #select rows and columns
        rows = [True] * m
        rows[i] = False
        cols = [True] * m
        cols[j] = False

        #compute minors matrix
        minors[i,j] = np.linalg.det(A[rows,:][:,cols])

        #compute H matrix
        H[i,j] = (-1) ** (i+j)

#cofactors matrix
C = H * minors

## Adjugate matrix
Ainv = C.T/np.linalg.det(A)
print(A@Ainv)
```

## Code challenge: gramschmit algorithm

```python
#Implement the G-S procedure!
#start with a square matrix, compute Q
#check Q'Q = 1
#check against np.linalg.qr()
m = 4
n = 4
A = np.random.randn(m,n)
Q = np.zeros((m,n))

#loop over columns
for i in range(n):
    Q[:,i] = A[:,i]
    a = A[:,i]
    #step1: orthagonalize the ith column in Q,
    #       relative to previous columns in A
    for j in range(i):
        q = Q[:,j]
        Q[:,i] = Q[:,i] - np.dot(a,q)/np.dot(q,q)*q
    #step2: normalize the column of Q
    Q[:,i] = Q[:,i]/np.linalg.norm(Q[:,i])
```

```python
import numpy as np
import scipy.linalg
import matplotlib.pyplot as plt

# Define matrices
A = np.array([[3, 2], [1, 3]])
B = np.array([[1, 1], [4, 1]])

# Generalized Eigenvalue Decomposition (GED)
eigvals, eigvecs = scipy.linalg.eig(A, B)

# Matrix-vector multiplication
Av = A @ eigvecs[:, 1]
Bv = B @ eigvecs[:, 1]
BinvAv = np.linalg.inv(B) @ A @ eigvecs[:, 1]

# Plot Av
plt.plot([0, eigvecs[0, 1]], [0, eigvecs[1, 1]], 'k', linewidth=4, label='$v_2$')
plt.plot([0, Av[0]], [0, Av[1]], 'r--', linewidth=2, label='$Av_2$')
plt.xlim([-3, 3]), plt.ylim([-3, 3])
plt.plot([-3, 3], [0, 0], 'k:')
plt.plot([0, 0], [-3, 3], 'k:')
plt.legend()
plt.title('Av')
plt.show()

# Plot Bv
plt.plot([0, eigvecs[0, 1]], [0, eigvecs[1, 1]], 'k', linewidth=4, label='$v_2$')
plt.plot([0, Bv[0]], [0, Bv[1]], 'r--', linewidth=2, label='$Bv_2$')
plt.xlim([-3, 3]), plt.ylim([-3, 3])
plt.plot([-3, 3], [0, 0], 'k:')
plt.plot([0, 0], [-3, 3], 'k:')
plt.legend()
plt.title('Bv')
plt.show()

# Plot B^{-1}Av
plt.plot([0, eigvecs[0, 1]], [0, eigvecs[1, 1]], 'k', linewidth=4, label='$v_2$')
plt.plot([0, BinvAv[0]], [0, BinvAv[1]], 'r--', linewidth=2, label='$B^{-1}Av_2$')
plt.xlim([-3, 3]), plt.ylim([-3, 3])
plt.plot([-3, 3], [0, 0], 'k:')
plt.plot([0, 0], [-3, 3], 'k:')
plt.legend()
plt.title('B$^{-1}$Av')
plt.show()
```

## Code challenge: SVD vs eigendecomposition for square symmetric matrices

```python
import numpy as np
import matplotlib.pyplot as plt

# Create a symmetric matrix (5x5)
A = np.random.randn(5, 5)
A = A.T @ A  # Make it symmetric

# Compute eigenvalues (L) and eigenvectors (W) and SVD (U, S, V)
L, W = np.linalg.eig(A)
U, s, V = np.linalg.svd(A)

# Sort eigenvalue outputs
sidx = np.argsort(L)  # Sorting indices for eigenvalues
L = L[sidx]  # Sort eigenvalues
W = W[:, sidx]  # Sort eigenvectors accordingly

# Images of all matrices
fig, ax = plt.subplots(2, 3, figsize=(10, 7))

# Eigen decomposition visualization
ax[0, 0].imshow(W)
ax[0, 0].set_title('W (eig)')

ax[0, 1].imshow(np.diag(L))
ax[0, 1].set_title('L (eig)')

ax[0, 2].imshow(np.zeros((1, 1)))  # Placeholder (optional)

# SVD visualization
ax[1, 0].imshow(U)
ax[1, 0].set_title('U (svd)')

ax[1, 1].imshow(np.diag(s))
ax[1, 1].set_title('$\sigma$ (svd)')

ax[1, 2].imshow(V)
ax[1, 2].set_title('V$^T$ (svd)')

plt.show()
```
